This code only does the Hamiltonian Simulation of a Tridiagonal Matrix. To see the entire code that solves a tridiagonal system go to the tridiagonal solver file.


```python
# Import the relevant libraries etc.

import cirq
from cirq.contrib.svg import SVGCircuit
from cirq.circuits import InsertStrategy


import sympy





# Set up the problem

a = 1
b = 2

# Problem Size
n_b = 5

# Number of Trotter Steps
m = 1






# Eigenvalue register
q_b = []
for q in range(n_b):
    q_b.append(cirq.LineQubit(q))


# Create the empty circuit
Ham_Sim_Circuit = cirq.Circuit()


# Check to see if all the qubits have been assigned to the correct registers
print(q_b)






# Implement H1. This can be done in one go and does not need to be split into Trotter Steps since it commutes with everything.
Ham_Sim_Circuit.append([
        cirq.X(q_b[n_b-1]),
        (cirq.Z(q_b[n_b-1])**(sympy.Symbol('a')*sympy.Symbol('t')/sympy.pi)),
        cirq.X(q_b[n_b-1]),
        ((cirq.Z(q_b[n_b-1])**(sympy.Symbol('a')*sympy.Symbol('t')/sympy.pi)))
    ],strategy=InsertStrategy.NEW)
         
         
# Implement a piece of H2 (second term)
Ham_Sim_Circuit.append(cirq.X(q_b[n_b-1])**(2*sympy.Symbol('b')*sympy.Symbol('t')/sympy.pi/sympy.Symbol('m')/2),strategy=InsertStrategy.NEW)





# This is the Trotter Formula Loop.

for i in range(m):
    # Implement a piece of H2 (third term)
    Ham_Sim_Circuit.append(cirq.X(q_b[n_b-1])**(-sympy.Symbol('b')*sympy.Symbol('t')/sympy.pi/sympy.Symbol('m')/2),strategy=InsertStrategy.NEW)
    
    # Implement a piece of H3
    # This is the H3 loop.
    # This is the loop over Cj
    for j in range(1,n_b):
        # Here we construct each individual Cj
        
        # First identify the top and bottom qubit for this Cj
        bottom = n_b-1
        top = n_b-1 - j
        
        # Now apply the first operation the controlled not gate
        Ham_Sim_Circuit.append(cirq.CNOT(q_b[top+1],q_b[top]),strategy=InsertStrategy.NEW)
        
        # Now apply a not gate since we want to apply a series of not controlled nots
        Ham_Sim_Circuit.append(cirq.X(q_b[top+1]),strategy=InsertStrategy.NEW)
        
        # Start the loop to insert the bunch of not controlled nots
        for k in range(top+2, bottom + 1):
            # Apply the controlled not operations
            Ham_Sim_Circuit.append(cirq.CNOT(q_b[top + 1], q_b[k]),strategy=InsertStrategy.NEW)
        
        
        # Restore the qubit that we applied a not operation to 
        Ham_Sim_Circuit.append(cirq.X(q_b[top+1]),strategy=InsertStrategy.NEW)
        
        
        # Here we apply the gate with multiple controls
        gate = (cirq.X(q_b[top + 1])**(-sympy.Symbol('b')*sympy.Symbol('t')/sympy.pi/sympy.Symbol('m')/2)).controlled_by(q_b[top])
        for k in range(top + 2, bottom + 1):
            gate = gate.controlled_by(q_b[k])
        Ham_Sim_Circuit.append(gate, strategy=InsertStrategy.NEW)
        
        
        # Apply a not gate again since we want to apply a series of not controlled nots
        Ham_Sim_Circuit.append(cirq.X(q_b[top+1]),strategy=InsertStrategy.NEW)
        
        # Start the loop to insert the bunch of not controlled nots
        for k in reversed(range(top+2, bottom + 1)):
            # Apply the controlled not operations
            Ham_Sim_Circuit.append(cirq.CNOT(q_b[top + 1], q_b[k]),strategy=InsertStrategy.NEW)
        
        
        # Restore the qubit that we applied a not operation to 
        Ham_Sim_Circuit.append(cirq.X(q_b[top+1]),strategy=InsertStrategy.NEW)
        
        
        # Apply the last operation, the controlled not gate
        Ham_Sim_Circuit.append(cirq.CNOT(q_b[top+1],q_b[top]),strategy=InsertStrategy.NEW)
        


    





# Implement a piece of H2 (last term)
Ham_Sim_Circuit.append(cirq.X(q_b[n_b-1])**(-2*sympy.Symbol('b')*sympy.Symbol('t')/sympy.pi/sympy.Symbol('m')/2),strategy=InsertStrategy.NEW)





print(Ham_Sim_Circuit)
SVGCircuit(Ham_Sim_Circuit)
```
